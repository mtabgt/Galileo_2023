### 1) ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
import numpy as np
array=np.array([True, False])
randomArray=np.random.choice(array,size=1000)
print("Tamaño del arreglo es de ",randomArray.size)
print("Tamaño en bytes de cada artículo: ",randomArray.itemsize)
print("Memoria total: ",randomArray.size*randomArray.itemsize, " bytes")
```

##### EL tamaño de 1000 elementos booleanos es de 1000 bytes.

### 5) Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
poblacion.query("Provincia=='Granada'").groupby(['Provincia']).agg(
    media_Granada = ('Poblacion', np.mean),
    mediana_Granada = ('Poblacion', np.median),
    desviacion_Granada = ('Poblacion', np.std),
    max_Granada = ('Poblacion', np.max),
    min_Granada = ('Poblacion', np.min),
)
```

### 6) Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
provincias=poblacion.drop_duplicates(['Provincia'])
for prov in provincias['Provincia']:
    plt.title(f'Histograma de población por municipio: {prov}')
    plt.bar('Municipio','Poblacion', width=1.0, data=poblacion[poblacion['Provincia'] ==prov])
    plt.xticks(rotation=45)
    plt.show()


```
